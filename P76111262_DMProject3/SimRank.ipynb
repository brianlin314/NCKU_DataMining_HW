```python
import networkx as nx
```

```python
f = open('./hw3dataset/graph_3.txt','r')
input = []
for line in f.readlines():
    s = line.split('\n')[0]
    s = s.split(',')
    s = tuple(s)
    input.append(s)
f.close
```

```python
G = nx.DiGraph()
G.add_edges_from(input)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(8, 8))
nx.draw_networkx(G, with_labels = True)
plt.show()
```
